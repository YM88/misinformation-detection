```python
import pandas as pd
from pycaret.classification import *
```

```python
training = pd.read_pickle('../data/te_sg_df_training.pickle')
testing = pd.read_pickle('../data/te_sg_df_testing_extra.pickle')
```

```python
cols = [col for col in training.columns.to_list() if col not in ['tweet', 'category']]
```

```python
training = training[cols]
testing = testing[cols]
```

```python
exp = setup(data=training, target='label', train_size=0.8, fold=5)
```

```python
compare_models()
```

```python
model = create_model('lightgbm')
```

```python
model = tune_model(model, optimize = 'F1')
```

```python
evaluate_model(model)
```

```python
from sklearn.metrics import classification_report
```

### Testing Performance

```python
y_score = predict_model(model, data=testing).Score
y_pred = predict_model(model, data=testing)['Label']
y_true = testing.label
```

```python
print(classification_report(y_true, y_pred, labels=None, target_names=None))
```
